# Graphs

```python
lambdas = np.linspace(UP[0,0], UP[0,-1], np.shape(UP)[1])

#ploted_part = (lambdas > 1091) & (lambdas < 1096)

#Full spectrum. This must be commented to plot just the part defined above
ploted_part = np.ones(np.shape(UP)[1], dtype = bool)
```

```python
def plot(xy, label, c=1):
    plt.plot(xy[0,ploted_part], c*xy[1,ploted_part], label=label)
```

```python
plt.figure(dpi=100, figsize=(16,9))
plt.xlabel("$\lambda$, cm$^{-1}$")
plt.title("Pure Components")

plot(UP, "UP")
plot(PEG, "PEG")
plot(TMAO, "TMAO")

plt.grid(True)
plt.legend()
plt.savefig('Graphs/Pure Components.jpg')
```

```python
plt.figure(dpi=100, figsize=(16,9))
plt.xlabel("$\lambda$, cm$^{-1}$")
plt.title("Above Binodal")

plot(TMAO_18_PEG_54_down, "TMAO 18%, PEG 54%, down")
plot(TMAO_18_PEG_54_up, "TMAO 18%, PEG 54%, up")
plot(TMAO_19_PEG_57_down, "TMAO 19%, PEG 57%, down")
plot(TMAO_19_PEG_57_up, "TMAO 19%, PEG 57%, up")
plot(TMAO_20_PEG_60_down, "TMAO 20%, PEG 60%, down")
plot(TMAO_20_PEG_60_up, "TMAO 20%, PEG 60%, up")


plt.grid(True)
plt.legend()
plt.savefig('Graphs/Above Binodal.jpg')
```

```python
plt.figure(dpi=100, figsize=(16,9))
plt.xlabel("$\lambda$, cm$^{-1}$")
plt.title("Below Binodal")

plot(TMAO_5_PEG_15, "TMAO 5%, PEG 15%")
plot(TMAO_10_PEG_30, "TMAO 10%, PEG 30%")
plot(TMAO_13_PEG_39, "TMAO 13%, PEG 39%")

plt.grid(True)
plt.legend()
plt.savefig('Graphs/Below Binodal.jpg')
```
