# Solution calcuations

```python
def concentrations(TMAO_f, PEG_f, NaPB_f=0.01, TMAO_i=60, PEG_i=80, NaPB_i=0.5, V=1000):
    TMAO = V * TMAO_f/TMAO_i
    PEG = V * PEG_f/PEG_i
    NaPB = V * NaPB_f/NaPB_i
    UP = V - TMAO - PEG - NaPB
    print('\nTMAO', TMAO_f, '%, PEG', PEG_f, '%, V =', V, 'mcl\n')
    if UP >= 0:
        print("TMAO:   ", TMAO, "mcl")
        print("PEG-600:", PEG, "mcl")
        print("UP:     ", UP, "mcl")
        print("NaPB:   ", NaPB, "mcl")
    else:
        TMAO = V * TMAO_f/100
        UP = V - TMAO - PEG - NaPB
        TMAO = TMAO/1000
        print("TMAO dry:", TMAO, "g")
        print("PEG-600: ", PEG, "mcl")
        print("UP:      ", UP, "mcl")
        print("NaPB:    ", NaPB, "mcl")
    print('\n')
```

```python
concentrations(TMAO_f=5, PEG_f=15)
concentrations(TMAO_f=10, PEG_f=30)
concentrations(TMAO_f=13, PEG_f=39)
concentrations(TMAO_f=18, PEG_f=54)
concentrations(TMAO_f=19, PEG_f=57)
concentrations(TMAO_f=20, PEG_f=60)
```

```python
n_TMAO = [30, 40, 50, 58, 75]
v_TMAO = [1000, 1000, 500, 398, 500]
for n, v in zip(n_TMAO, v_TMAO):
    concentrations(TMAO_f=n, PEG_f=0, V=v)
```

```python
n_PEG = [65, 70, 75, 85, 90, 95]
for n in n_PEG:
    concentrations(TMAO_f=0, PEG_f=n, PEG_i=100)
```
